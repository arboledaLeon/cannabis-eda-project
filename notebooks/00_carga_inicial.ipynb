### 🧠 Un poco de teoría antes de empezar

- **DataFrame:** es la estructura central de pandas. Piénsalo como una **hoja de cálculo** dentro de Python:
  filas (cada cepa) y columnas (cada característica).
- **¿Por qué cargamos desde un CSV local y no desde internet?** El dataset original vive en Hugging Face,
  pero la librería para descargarlo (`datasets`) está rota en Python 3.13. Además, cargar desde un archivo
  local es **más rápido y reproducible**: cualquiera que clone el repo obtiene exactamente los mismos datos.

El código de carga vive en `src/data_loading.py` para mantener este notebook limpio.

## 1. Importar herramientas y cargar los datos

```python
import sys
sys.path.append("..")   # permite que Python "vea" la carpeta src/ (que está un nivel arriba)

import pandas as pd
from src.data_loading import load_raw

# Mostrar todas las columnas cuando imprimamos el DataFrame (por defecto oculta algunas)
pd.set_option("display.max_columns", 40)

df = load_raw()
print(f"Filas (cepas):     {df.shape[0]}")
print(f"Columnas (rasgos): {df.shape[1]}")
```

### Otras formas de mirar los datos

No basta con ver el principio. Conviene mirar también el **final** y algunas filas **al azar**, por si los
datos cambian de forma a lo largo del archivo (algo común en datos scrapeados).

`df.tail()` muestra las **últimas** filas:

```python
df.tail()
```

`df.sample(3)` toma 3 filas **al azar**. Es útil para no dejarnos engañar por el orden del archivo
(cada vez que lo ejecutes saldrán filas distintas):

```python
df.sample(3)
```

También podemos mirar **solo las columnas que nos interesan**, poniendo sus nombres entre corchetes.
Por ejemplo, el nombre de la cepa junto con su productor (breeder):

```python
df[["strain_name", "breeder"]].head(10)
```
